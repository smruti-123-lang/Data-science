```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

#create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("basic line plot")
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2, markersize=9)
plt.grid(True)
```

```python
##Multiple plots
## sample data
x = [1,2,3,4,5]
y1= [1,4,9,16,25]
y2= [1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2,color='green')
plt.title("plot 2")

```

```python
## bar plot
categories = ['A','B','C','D','E']
values=[95,107,88,99,100]

plt.bar(categories,values,color='purple')
```

```python
##histogram = for the distribution of data sets

data = [1,2,3,4,4,4,5,5,3,6,6,7,6]
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

```

```python
## scatter plot

x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y,color="blue",marker='x')
```

```python
##pie chart 
labels =['A','B','C','D']
sizes = [30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightblue']
explode=(0.1,0,0,0)


plt.pie(sizes,labels = labels,explode=explode,colors = colors,autopct="%1.1f%%",shadow = True)
```
