```python
import numpy as np

# Create 4x4 array with values 1 to 16
arr = np.arange(1, 17).reshape(4, 4)

# Replace diagonal elements with 0
np.fill_diagonal(arr, 0)

print(arr)
```

```python
import numpy as np

# Create a 6x6 array with values from 1 to 36
arr = np.arange(1, 37).reshape(6, 6)

# Extract sub-array: rows 3–5 and columns 2–4 (indexing starts from 0)
sub_arr = arr[2:5, 1:4]

print("Original Array:\n", arr)
print("\nExtracted Sub-array:\n", sub_arr)
```
