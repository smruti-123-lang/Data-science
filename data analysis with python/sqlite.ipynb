```python
import sqlite3
```

```python
#connect to an SQLite database
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
###create a table
cursor.execute('''
Create table if not exists employees(
                id integer Primary Key,
                name Text Not Null,
                age Integer,
                Department text
                )
            ''')
## commit the changes
connection.commit()
```

```python
##insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
                values("Krish",32,"data scientist")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("nish",32,"data scientist")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("bob",32,"data scientist")

''')
connection.commit()
```

```python
## Query the data from sqlite
cursor.execute('Select * from employees')
rows = cursor.fetchall()

##print the queried data
for row in rows:
    print(row)
```

```python
cursor.execute('''
UPDATE employees
Set age = 34
where name = "Krish"
''')
connection.commit()
```

```python

cursor.execute('Select * from employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
##delete the data
cursor.execute('''
DELETE from employees
               where name = "nish"
''')

connection.commit()
```

```python
connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT
)
''')


```

```python
cursor.executemany('''
INSERT INTO sales (date, product, sales, region)
VALUES (?, ?, ?, ?)
''', [
    ('2025-01-03', 'Laptop', 12, 'North'),
    ('2025-01-03', 'Smartphone', 25, 'South'),
    ('2025-01-04', 'Tablet', 9, 'East'),
    ('2025-01-04', 'Headphones', 30, 'West'),
    ('2025-01-05', 'Laptop', 15, 'South'),
    ('2025-01-05', 'Smartwatch', 7, 'North'),
    ('2025-01-06', 'Tablet', 11, 'West'),
    ('2025-01-06', 'Laptop', 20, 'East'),
    ('2025-01-07', 'Smartphone', 18, 'North'),
])

```

```python
cursor.execute('Select * from sales')
rows = cursor.fetchall()
for row in rows:
    print(row)
```
