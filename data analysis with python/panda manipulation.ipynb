```python
##filling missing values with mean

df['Crop_Year_filNa'] = df['Crop_Year'].fillna(df['Crop_Year'].mean())
df
```

```python
## rename the columns
df = df.rename(columns={'Crop': 'Phasal'})
df.head(1)
```

```python
## change datatypes
df['jaga']=df['Area'].astype(int)
df.head(1)
```

```python
df['new_year']=df['Area'].apply(lambda x:x*2)
df.head(2)
```

```python
grouped_mean=df.groupby('Crop_Year')['Yield'].mean()
print(grouped_mean)
```

```python
df.groupby('Phasal')['Yield'].mean()
```

```python
df.groupby(['Phasal', 'Crop_Year'])['Yield'].sum()

```

```python
##aggregate multiple functions
grouped_agg = df.groupby('Phasal')['Yield'].agg(['mean','sum','count'])
grouped_agg.head(3)
```

```python
##merging and joining datafra
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value2': [4, 5, 6]})
df1
```

```python
df2
```

```python
merged_inner = pd.merge(df1, df2, on='key', how='inner')
merged_inner
```

```python
merged_outer = pd.merge(df1, df2, on='key', how='outer')
merged_outer
```

```python
merged_left = pd.merge(df1, df2, on='key', how='left')
merged_left 
```

```python
merged_right = pd.merge(df1, df2, on='key', how='right')
merged_right
```
