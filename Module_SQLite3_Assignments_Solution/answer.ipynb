```python
connection = sqlite3.connect('commontest.db')
```

```python
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees(
id integer primary key,
name text,
age integer,
department text
)''')
```

```python
cursor.execute('''
Insert Into employees(name,age,department)
                values("Krish",32,"data scientist")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("nish",32,"data scientist")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("bob",32,"data scientist")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("knbob",32,"data engineer")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("bouob",32,"ai scientist")

''')
cursor.execute('''
Insert Into employees(name,age,department)
                values("boob",32,"ml scientist")

''')
connection.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('Select * from employees where department="data scientist"')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
UPDATE employees 
SET age = 33 
WHERE department="data scientist"
''')
connection.commit()
cursor.execute('Select * from employees where department="data scientist"')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
DELETE FROM employees 

WHERE department="data engineer"
''')
connection.commit()
cursor.execute('Select * from employees ')
rows = cursor.fetchall()

for row in rows:
    print(row)
```
